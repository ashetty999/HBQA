```python
df_prompt_with_qa= pd.read_csv('Question_Prompts_with_QA.csv')
df_prompt_with_qa= df_prompt_with_qa[df_prompt_with_qa.QA.notna()]
df_prompt_with_qa
```

```python
df_prompt_with_qa.shape
```

```python
df_qa_final = pd.DataFrame(columns=['QuesId','SectionId','ContextId','Question','Answer','Reference']) 

qtn_num=0
for prompt_qa_row in df_prompt_with_qa.index:

    content = df_prompt_with_qa.loc[prompt_qa_row,'QA']
    ContextId = df_prompt_with_qa.loc[prompt_qa_row, 'Context_Id']
    SectionId =  df_prompt_with_qa.loc[prompt_qa_row, 'Section_Id']


    print('Section:', SectionId, "ContextId", ContextId)
    qa_txt = content.split('\n')
    for i in range(len(qa_txt)):
        para=qa_txt[i]
        if para!='': 
            print(para)
            if para[0:8]=='Question':
                Question = para 
                try:
                    Answer = qa_txt[i+1]

                    if qa_txt[i+2]!="":
                        Reference = qa_txt[i+2]
                    else:
                        Reference = ""

                    qtn_num += 1
                    qa_final_row = qtn_num, SectionId, ContextId, Question, Answer, Reference
                    df_qa_final.loc[qtn_num] = qa_final_row
                    
                    print(Question)
                except:
                    pass
```
