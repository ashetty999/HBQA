```python
# from google.colab import drive
# drive.mount('/content/drive')

apipath = r'H:\\My Drive\\config\\hbqa.txt'
import configparser

config = configparser.ConfigParser()
config.read(apipath)
secret_key = config['global']['OPENAI_API_KEY']
datapath = config['global']['DATA_FOLDER']
```

```python
df= pd.read_csv(datapath+'04-Chunk_with_Prompts.csv')
print(df.shape)
df.head(5)
```

```python
import openai 
```

```python
import openai
openai.api_key = secret_key
def get_qa(qa_prompt):
    try:
        response = openai.Completion.create(
            model="text-davinci-003",
            prompt=qa_prompt,
            max_tokens=1000,
            temperature=0,
            frequency_penalty=1
        )
        return response['choices'][0]['text']
    except:
        print("Error")
        return ""
```

```python
start=400
end=450

for row in range(end-start):
    qa_prompt = df.loc[row,'Prompt']
    response = get_qa(qa_prompt)
    df.loc[row+start,"QA"] = response 
    print("Row: ", row+start)

#response = df['Prompt'].apply(get_qa)
```

```python
df=df.loc[df['QA'].notna(),:]
```

```python
qa_file="05-QA_Chunk"+str(start)+"_"+str(end-1)+".csv"
```

```python
df.to_csv(datapath+qa_file)
```
